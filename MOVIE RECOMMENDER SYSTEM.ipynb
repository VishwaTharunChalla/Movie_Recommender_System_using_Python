```python
def content_based_filtering(movie_title,no_of_recos):
    check_genre=merged_dataset[merged_dataset.title==movie_title]
    check_unique_genre=check_genre.genres.unique()[0]
    abc=genres_based.groupby(['title','genres'])['movieId'].count().reset_index().sort_values('movieId',ascending=False)
    abc_set_index=abc.set_index('title')
    abc_set_index.drop([movie_title],axis=0,inplace=True)
    v=abc_set_index.drop(columns=['movieId','genres']).reset_index()[:no_of_recos]
    return v
```

## Ratings Matrix
## pivot_table():- It is a statistical tool which summarizes and reorganizes the selected columns and rows of data in a spreadsheet (or) database.
## pivot():- It is a function which is used for reshape a given dataframe organized by given index/column values.

```python
ratings_matrix=final_ratings.pivot_table(index='userId',
                                        columns='movieId',
                                        values='rating',
                                        aggfunc='mean',
                                        fill_value=0)
```

```python
ratings_matrix
```

## Cosine-Based Similarity:- It is the metric used to measure how similar the documents are irrespective of their size

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_of_choice=ratings_matrix[ratings_matrix.index==38]  # I took 38 index for example
user_of_choice
```

```python
others=ratings_matrix[ratings_matrix.index != 38]  # not equals to 38 index(remaining all)
others
```

```python
cosine_similarity(user_of_choice,others)
```

```python
similarity_array=cosine_similarity(user_of_choice,others).flatten() #flatten() returns the array into one dimension
similarity_array
```

```python
similarity_series=pd.Series(similarity_array,index=others.index)
similarity_series
```

```python
similarity_series.sort_values(ascending=False)[:10]
```

```python
top_similar=similarity_series.sort_values(ascending=False)[:10].index
top_similar
```

```python
def similar_users(user_no,k):    # k= no. of  similar userId's to show, user_no=userId from ratings_dataset
    user_of_choice=ratings_matrix[ratings_matrix.index==user_no]  
    others=ratings_matrix[ratings_matrix.index!=user_no]
    similarity_ser=pd.Series(cosine_similarity(user_of_choice,others).flatten(),index=others.index)
    top_similar=similarity_ser.sort_values(ascending=False)[:k].index
    return top_similar
```

```python
def recommendation(user_id, num_recommendations):
    similar_user_data = ratings_matrix[ratings_matrix.index.isin(similar_users(user_id, 20))]
    df_sim_user = pd.DataFrame(similar_user_data.apply(lambda x : x[x>0].mean()), columns = ['mean_ratings']).reset_index()
    user_of_choice = ratings_matrix.loc[ratings_matrix.index == user_id]
    user_data = pd.DataFrame(user_of_choice.T.values , columns = ['ratings'], index = user_of_choice.columns)
    
    unseen_movies = user_data[user_data.ratings == 0].index
    sorted_unseen = df_sim_user[df_sim_user.movieId.isin(unseen_movies)].sort_values('mean_ratings', ascending=False)[:num_recommendations]

    # Recommended movies 
    recos = pd.merge(movies_dataset, sorted_unseen)   # merge with movies_dataset only
    recos.drop(columns=['movieId','genres','mean_ratings'],inplace=True)
    final=recos.rename(columns={'title':'Movie Title'})
    return final
```
